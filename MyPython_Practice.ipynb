```python
# Counting special characters in a string
def count_sp_char(string):
    sp_char = "!@#%^&*()_+"
    count = 0
    for i in string:
        if i in sp_char:
            count += 1
    return count

text = "Hol@ !! How @re you?"
print(count_sp_char(text))
```

```python
def pyramid(n):
    for i in range(n):
        for j in range(i,n):
            print(" ", end="")
        for j in range(i+1):
            print("*", end="")
        for j in range(i):
            print("*", end="")
        print("")
        
pyramid(5)
```

```python
def isprime(num):
    for i in range(2,num):
        if num%i == 0:
            return False
    return True

def prime_generator(n):
    num = 2
    while(n):
        if isprime(num):
            yield num
            n -= 1
        num += 1

print(list(prime_generator(5)))
```

```python
# Decorators

def welcome(fx):
    def mfx(*t, **d):
        print("Before hello function")
        fx(*t,**d)
        print("Thanks for using the funxtion")
    return mfx

@welcome
def hello():
    print("Hello !")
    
hello()
```

```python
class Calculator:
    def __init__(self, func):
        self.function = func
    
    def __call__(self, *t, **d):
        result = self.function(*t, **d)
        return result**2
    
@Calculator
def add(a,b):
    return a+b

add(10,20)
```

```python
# Generators (yeild can be used as if the function returned something, but doesn't actually stop the flow)

def evenNumber(n):
    i = 1
    while n:
        yield 2*i
        i+=1
        n-=1

it = evenNumber(10)
print(type(it)) # Generator can wrapped around list to convert to list
#print(list(it))
#Another way of iterating 
even_list = []
while True:
    try:
        even_list.append(next(it))
    except StopIteration:
        break
print(even_list)
```

```python
# Difference  between class object and instance object

class test:
    x = 20
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def show(self):
        print(self.a, self.b)
    @staticmethod
    def show_values():
        print("This is a static method and can't access self variables")

test.show_values()
print(test.x)
t1 = test(4,5)
t1.show()
test.show_values()
```
